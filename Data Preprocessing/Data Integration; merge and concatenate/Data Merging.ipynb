# Data Merging:
Data merging involves combining datasets based on a common attribute or key. This is typically done when you have information spread across multiple tables or files, and you want to consolidate them into a single dataset.

```python
import pandas as pd
```

### Create dummy datasets

```python
df1 = pd.DataFrame({
    "Id":[11,22,33,44,55,66,77,88,99],
    "Name":["Ali","Faraz","Bilal","Faizan","Waseem","Haider","Abdullah khan","Azeem","Taha"],
    "Age":[23,22,20,24,23,18,19,25,26]
})
```

```python
df1
```

```python
df2 = pd.DataFrame({
    "Id":[11,22,33,44,55],
    "Occupation":["OOP Developer","Pool Engineer","Road Engineer","Doctor","Mechanic"]
})
```

```python
df2
```

### Now we will merge data

- how = "left"

```python
# left dataset ki baat sunni jaye gii
df = pd.merge(df1,df2, on="Id", how="left")
df
```

- how = "right"

```python
# right dataset ki baat sunni jaye gi
df = pd.merge(df1,df2, on="Id", how="right")
df
```

- how = "inner"

```python
# kamzoor ki baat sunni jaye gi
df = pd.merge(df1,df2, on="Id", how="inner")
df
```

- how = "outer"

```python
# taqat war ki baat sunni jaye gii
df = pd.merge(df1,df2, on="Id", how="outer")
df
```
